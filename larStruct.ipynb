# 

- **larApply**(Mat) $\rightarrow$generatore di funzione,  
questa funzione tramite la matrice affine data in input, esegue una trasformazione affine, e restituisce un generatore di funzione la nuova lista delle coordinate dei vertici e un vettore di celle.  

  
- **evalStruct**(struct) $\rightarrow$ ndarray,  
in base al tipo di ogni elemento di struct riempie l'array vuoto con una lista contenente: le coordinate dei vertici modificate attraverso una trasformazione affine e il vettore di celle.   



- **vcode**(int)(array) $\rightarrow$ string,  
approssima ogni elemento dell'array con precisione data dall'intero, poi trasforma l'array ottenuto in stringa.    


- **struct2lar**(struct) $\rightarrow$ ndarray,  

- **larRemoveVertices**(ndarray)(ndarray) $\rightarrow$ ndarray,  
prendi in input array di vertici e celle e li restituisce togliendo i vertici doppi.   


- **larEmbed**(int)(ndarray) $\rightarrow$ ndarray,  
se int > 0 aggiunge una dimensione a tutti gli elementi del primo array in ndarray, altrimenti restituisce l'ndarray non modificato.  


- **larBoundary**(struct) $\rightarrow$ ndarray  
applica **struct2lar** all'input restituendo un ndarray.
  
  
- **embedStruct**(int)(struct,string) $\rightarrow$ tupla
